```python
import tkinter as tk
# Tkinter is built in Python Library for GUI Application
# scrolledtext → special widget that gives a Text box with scroll bar
from tkinter import scrolledtext

# re → Regular Expression library used for pattern matching (text cleaning) Regex
import re

# Counter → used to count frequency of words efficiently
from collections import Counter

# ================= CLASS DEFINITION =================
# We are creating a CLASS because:
# - We want to bundle data (widgets) + behavior (functions) together
# - This is Object Oriented Programming (OOP)

class TextProcessorGUI:

    # ================= CONSTRUCTOR =================
    # __init__ is a special method called CONSTRUCTOR
    # It runs automatically when we create object: TextProcessorGUI(root)
    def __init__(self, root):

        # root → main window (Tk object passed from outside)
        # This is called dependency injection (we pass main window into class)

        self.root = root  # ATTRIBUTE → storing root window inside object

        # set title of window (appears on top bar)
        self.root.title("Text Processor")

        # set size of window (width x height)
        self.root.geometry("800x600")


        # ================= LABEL WIDGET =================
        # Label is a simple display widget (non-editable text)
        # It is attached to root window

        tk.Label(root, text="Enter Text:", font=("Arial", 12, "bold")).pack()

        # .pack() → geometry manager
        # It automatically places widget vertically one below another


        # ================= INPUT TEXT BOX =================
        # ScrolledText = Text area + scrollbar
        # Used for multi-line input

        self.input_box = scrolledtext.ScrolledText(
            root,        # parent window
            height=10,   # visible height of box
            wrap=tk.WORD # wraps text at word boundary (not mid-word)
           
        )

        # pack → places widget in GUI layout
        self.input_box.pack(fill=tk.BOTH, padx=10, pady=5)
        # it can be expanded in both x and y

        # ================= BUTTON FRAME =================
        # Frame = container widget (used to group buttons)

        btn_frame = tk.Frame(root)
        
        btn_frame.pack(pady=10)


        # ================= BUTTONS =================
        # Each button has:
        # - text (label shown on button)
        # - command (function to call when clicked)
        # bg - background color

        

        
        tk.Button(btn_frame, text="Lowercase", command=self.lowercase, bg="lightgreen").grid(row=0, column=0, padx=5)
        tk.Button(btn_frame, text="Uppercase", command=self.uppercase, bg="lightgreen").grid(row=0, column=1, padx=5)
        tk.Button(btn_frame, text="Remove Spaces", command=self.remove_spaces, bg="lightgreen").grid(row=0, column=2, padx=5)
        tk.Button(btn_frame, text="Remove Punctuation", command=self.remove_punct, bg="lightgreen").grid(row=0, column=3, padx=5)
        tk.Button(btn_frame, text="Title Case", command=self.title_case, bg="lightgreen").grid(row=0, column=4, padx=5)
        tk.Button(btn_frame, text="Reset Stats", command=self.reset_stats, bg="lightgreen").grid(row=0, column=5, padx=5)

        # NOTE:
        # grid() → another geometry manager (row-column system)
        # pack() and grid() should NOT be mixed in same container normally


        # ================= OUTPUT BOX =================
        # Another ScrolledText widget to show processed text output

        tk.Label(root, text="Output:", font=("Arial", 12, "bold")).pack()

        self.output_box = scrolledtext.ScrolledText(root, height=10, wrap=tk.WORD)
        self.output_box.pack(fill=tk.BOTH, padx=10, pady=5)


        # ================= STATS LABEL =================
        # Label to show dynamic stats like word count etc.

        self.stats_label = tk.Label(root, text="", font=("Arial", 11), fg="blue")
        self.stats_label.pack(pady=10)

        # initial call to show stats when program starts
        self.update_stats()


    # ================= HELPER METHOD =================
    # This method reads text from input box

    def get_text(self):
        # "1.0" → start of text widget (line 1, character 0)
        # tk.END → end of text widget
        return self.input_box.get("1.0", tk.END).strip()


    # ================= HELPER METHOD =================
    # This method writes text into output box

    def set_text(self, text):
        # delete everything in output box first
        self.output_box.delete("1.0", tk.END)

        # insert new text at beginning
        self.output_box.insert(tk.END, text)


    # ================= TEXT OPERATIONS =================

    def lowercase(self):
        # get text → convert → send to output
        text = self.get_text().lower()
        self.set_text(text)
        self.update_stats(text)

    def uppercase(self):
        text = self.get_text().upper()
        self.set_text(text)
        self.update_stats(text)

    def remove_spaces(self):
        # regex: \s+ → matches multiple spaces/tabs/newlines
        # replace them with single space
        text = re.sub(r'\s+', ' ', self.get_text()).strip()
        self.set_text(text)
        self.update_stats(text)

    def remove_punct(self):
        # regex explanation:
        # [^\w\s] → means NOT (word characters OR whitespace)
        # so it removes punctuation
        text = re.sub(r'[^\w\s]', '', self.get_text())
        self.set_text(text)
        self.update_stats(text)

    def title_case(self):
        text = self.get_text().title()
        self.set_text(text)
        self.update_stats(text)

    def reset_stats(self):
        # clears output box and stats label
        self.output_box.delete("1.0", tk.END)
        self.stats_label.config(text="")


    # ================= ANALYSIS METHOD =================

    def update_stats(self, text=None):

        # if no text passed, read from input box
        if text is None:
            text = self.get_text()

        # -------- WORD COUNT --------
        words = text.split()  # splits by whitespace
        word_count = len(words)

        # -------- CHARACTER COUNT --------
        char_count = len(text)

        # -------- SENTENCE COUNT --------
        # split based on . ! ?
# '''What It Breaks Down To
#      r       : Specifies a raw string so backslashes are treated literally.
#      [...]   :A character class that matches any single character inside the brackets.
#      .!?     : The specific characters to look for (a period, exclamation mark, or question mark).
#      +       : A quantifier meaning "one or more times" in a row".
# '''
        
        sentences = re.split(r'[.!?]+', text)
        sentence_count = len([s for s in sentences if s.strip()])


        # -------- MOST COMMON WORDS --------
        # extract words using regex \w+ (letters + digits + underscore)
        freq = Counter(re.findall(r'\w+', text.lower()))

        # top 3 frequent words
        most_common = freq.most_common(3)

        # -------- FINAL DISPLAY STRING --------
        stats = (
            f"Words: {word_count} | "
            f"Characters: {char_count} | "
            f"Sentences: {sentence_count}\n"
            f"Top Words: {most_common}"
        )

        # update label dynamically
        self.stats_label.config(text=stats)


# ================= MAIN PROGRAM =================
# This runs only when file is executed directly

if __name__ == "__main__":

    # Tk() → creates main GUI window (event loop base)
    root = tk.Tk()

    # object creation → constructor __init__ runs automatically
    app = TextProcessorGUI(root)

    # starts GUI event loop:
    # - listens for button clicks
    # - keeps window open
    root.mainloop()
```
